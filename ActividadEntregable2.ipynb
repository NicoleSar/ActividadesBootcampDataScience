*1. Crea un arreglo 5x3 de números aleatorios entre 0 y 100*

Uso del método de numpy *random.randint:* 

random.randint(low, high=None, size=None, dtype=int)

```python
array = np.random.randint(100, size = (5, 3))
array
```

*2. ¿Cuál es la suma total de los valores del arreglo?*

```python
print(f'Sumatoria total de los valores del arreglo: {np.sum(array)}')
```

*3.¿Cúal es el promedio de cada columna del arreglo?*

```python
print(f'Promedio de cada columna del arreglo: {np.mean(array,axis = 0)}')
```
