```python
#Count by SKU category: presentation forms (kegs, bottles, pints, cans) 
BreweryOp['SKU'].value_counts()
```

```python
#Simple filter: filter lager beer data
BreweryOp[BreweryOp ['Beer_Style'] == 'Lager']

#Multiple filter: filter SKU Kegs lager beer data (kegs format)
BreweryOp[(BreweryOp ['Beer_Style'] == 'Lager') & (BreweryOp ['SKU'] == 'Kegs')]
#Multiple filter: filter SKU Kegs lager beer data (kegs format) with a quality score greater than 9.0
BreweryOp[(BreweryOp ['Beer_Style'] == 'Lager') & (BreweryOp ['SKU'] == 'Kegs') & (BreweryOp ['Quality_Score'] >= 9.0)]
#Multiple filter: filter SKU Kegs lager beer data (kegs format) with a quality score greater than 9.0 and Bottling/Kegging Loss (%) lower than 2.0
BreweryOp[(BreweryOp ['Beer_Style'] == 'Lager') & (BreweryOp ['SKU'] == 'Kegs') & (BreweryOp ['Quality_Score'] >= 9.0) & (BreweryOp ['Loss_During_Bottling_Kegging'] <= 2.0)]

#Creating a new dataset applying previous filters 
BreweryOpLagerKegs = BreweryOp[(BreweryOp ['Beer_Style'] == 'Lager') & (BreweryOp ['SKU'] == 'Kegs') & (BreweryOp ['Quality_Score'] >= 9.0) & (BreweryOp ['Loss_During_Bottling_Kegging'] <= 2.0)]
BreweryOpLagerKegs.head()
```
